```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
from fairlearn.metrics import MetricFrame, selection_rate
```

```python
data_path = Path(os.getcwd()).parent / "data" / "dataset_diabetes"
df = pd.read_csv(data_path / "diabetic_prepared.csv")
```

```python
y_true = (df.readmit_30_days == True) * 1
gender = df["gender"]
race = df["race"]
age = df["age"]
```

Data bias based on gender

```python
selection_rates = MetricFrame(
    metrics=selection_rate, 
    y_true=y_true, 
    y_pred=y_true, 
    sensitive_features=gender
)

fig = selection_rates.by_group.plot.bar(
    legend=False, rot=0, title="Fraction readmitted within 30 days"
)
```

Data bias based on race

```python
selection_rates = MetricFrame(
    metrics=selection_rate,
    y_true=y_true, 
    y_pred=y_true, 
    sensitive_features=race
)

fig = selection_rates.by_group.plot.bar(
    legend=False, rot=0, title="Fraction readmitted within 30 days"
)
```
