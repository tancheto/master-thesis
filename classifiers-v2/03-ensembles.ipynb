XGBoost

```python
xgb = XGBClassifier(
    subsample=0.75,
    min_child_weight=10,
    max_depth=6,
    learning_rate=0.01,
    gamma=1,
    colsample_bytree=1.0,
    booster='gbtree',
    objective='binary:hinge',
    silent=True,
    nthread=-1
)

train_model_on_datasets(xgb)
```
